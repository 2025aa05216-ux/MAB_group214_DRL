```python
# ----------------------------
# Configuration
# ----------------------------

GRID_SIZE = 5
MAX_BATTERY = 15
WIND_PROB = 0.20
THETA = 1e-3
GAMMA = 0.95

# TODO:
# Define locations based on your student ID
#START = (0, 0)
grid = [
    ["S","F","F","D","F"],
    ["F","X","F","F","F"],
    ["R","F","W","D","F"],
    ["F","C","F","X","F"],
    ["F","D","F","R","F"]
]


```

```python
class DroneRescueEnv:

    def __init__(self):

        self.grid = grid

        self.start = (0,0)

        self.rescue_locations = [(2,0),(4,3)]

        self.charging_station = (3,1)

        self.danger_cells = [
            (0,3),
            (2,3),
            (4,1)
        ]

        self.blocked_cells = [
            (1,1),
            (3,3)
        ]

        self.wind_cell = (2,2)

        self.reset()

    def reset(self):

        self.position = self.start

        self.battery = MAX_BATTERY

        self.rescued = [False,False]

        self.steps = 0

        return self.get_state()

    def get_state(self):

        return (
            self.position,
            self.battery,
            tuple(self.rescued)
        )
```

```python
ACTIONS = {
    "UP":(-1,0),
    "DOWN":(1,0),
    "LEFT":(0,-1),
    "RIGHT":(0,1),
    "HOVER":(0,0)
}
```

```python
def valid_position(self,row,col):

    if row < 0 or row >= GRID_SIZE:
        return False

    if col < 0 or col >= GRID_SIZE:
        return False

    if (row,col) in self.blocked_cells:
        return False

    return True
```

```python
def step(self,action):

    reward = -1

    self.steps += 1

    # battery usage

    self.battery -= 1

    row,col = self.position

    # wind behaviour

    if self.position == self.wind_cell:

        if random.random() < WIND_PROBABILITY:

            action = random.choice(
                ["UP","DOWN","LEFT","RIGHT"]
            )

    dr,dc = ACTIONS[action]

    new_row = row + dr
    new_col = col + dc

    if self.valid_position(new_row,new_col):

        self.position = (new_row,new_col)

    # charging station

    if self.position == self.charging_station:

        reward += 5

        self.battery = MAX_BATTERY

    # rescue targets

    for idx,target in enumerate(
            self.rescue_locations):

        if (self.position == target and
            self.rescued[idx] == False):

            reward += 20

            self.rescued[idx] = True

    # danger zones

    if self.position in self.danger_cells:

        reward -= 10

    done = False

    if self.battery <= 0:

        reward -= 20

        done = True

    if all(self.rescued):

        done = True

    if self.steps >= MAX_STEPS:

        done = True

    return (
        self.get_state(),
        reward,
        done
    )
```

```python
states = []

for row in range(GRID_SIZE):

    for col in range(GRID_SIZE):

        if (row,col) in [(1,1),(3,3)]:
            continue

        for battery in range(MAX_BATTERY+1):

            for rescue_state in range(4):

                rescued1 = bool(rescue_state & 1)
                rescued2 = bool(rescue_state & 2)

                states.append(
                    (
                        (row,col),
                        battery,
                        (rescued1,rescued2)
                    )
                )

print("Total States :",len(states))
```

```python
start_time = time.time()

iteration = 0

while True:

    delta = 0

    iteration += 1

    for state in states:

        values = []

        for action in ACTIONS.keys():

            values.append(
                np.random.random()
            )

        best_value = max(values)

        delta = max(
            delta,
            abs(best_value - V[state])
        )

        V[state] = best_value

    if delta < theta:
        break

runtime = time.time() - start_time

print("Iterations :",iteration)
print("Final Delta :",delta)
print("Runtime :",runtime)
```
